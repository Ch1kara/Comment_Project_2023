```python
import csv
with open('Class_info.csv') as file: #Opens the csv file containing the student data and inputs it into a list of lists
    HEADER = next(file)
    INFO = []
    data = csv.reader(file)
    for line in data:
        INFO.append(line)


def class_intro(student):
    '''Takes in the name of a student and outputs a personalized summary of Precalculus with Proofs,
    change the f string as needed based on your classes description'''
    return(f"\t{student}, this semester in Pre-Calculus with Proofs, I have had the privilege to work with you as we dive deeper into the study of trigonometric identities and dip our toes into various foundations of differential calculus. We started the year by reviewing the definitions of sine and cosine functions using right triangles and then used the unit circle to extend the domain of those functions to all real numbers. We then used transformations of the plane to prove several trigonometric identities involving sine and cosine functions. Applications included proving de Moivre's formula and finding nth roots of unity. We also proved and applied the law of sines, the law of cosines, and Heron’s area formula. The concept of inverse functions was used to study inverse trigonometric functions and to solve trigonometric equations. We were also introduced to sequences and the notion of the limit of a sequence from both an intuitive and rigorous approach.")


def strong_trait(student):
    '''Takes in the name of a student and outputs a personalized description
    of their strong trait and how they have demonstrated said trait throughout the class'''
    for i in INFO:
        if i[0] == student:
            strong_trait = i[2]
            reason = i[5]
            break
    return(f"\n\t{student}, I've noticed that you're extremely {strong_trait.lower()} during class. You constantly show this through your ability to {reason}. I encourage you to continue being {strong_trait.lower()} throughout the rest of our time together.")

def weakness(student):
    '''Takes in the name of a student and outputs a personalized description of their area for improvement
    and why that is an area for improvement'''
    for i in INFO:
        if i[0] == student:
            weak_trait = i[3]
            reason = i[4]
            break
    return(f"Looking to the second semester, one thing you could work on is your {weak_trait.lower().strip()} in class. I've noticed that you {reason}. I recommend that you continue to work on your {weak_trait.lower()} in class.")


def learning_outcome(student):
    '''Takes in the name of a student and outputs a personalized description
     of one learning outcome they are strong in and one they can improve on'''
    for i in INFO:
        if i[0] == student:
            strong_outcome = i[6]
            weak_outcome = i[7]
        if i[1] in ['A','A-','B+']:
            level = 'On another note, you have demonstrated excellence with many of the learning outcomes in this class.'
        elif i[1] in ['B','B-','C+','C','C-','D+']:
            level = 'On another note, you have shown strength in a few of the learning outcomes in this class.'
    return(f"\n\t{level} One I would like to highlight is how you can consistently {strong_outcome}. You consistently demonstrate your excellence in this skill and I look forward to your continued success in this area. However, I've noticed that you could work on your ability to {weak_outcome}. I am always available for one-on-one assistance during office hours if you feel you need extra help with this skill.")


def strong_test(student):
    '''Takes in the name of a student and outputs a personalized description
     of their best test and the reason why it was their best'''
    for i in INFO:
        if i[0] == student:
            strong_test = i[8]
            test_score = i[9]
            reason = i[10]
    if int(test_score) > 92:
        add_on = 'You should be proud of your work.'
    try:
        return(f"\n\tThis semester, we have had three exams. You had the most success on the {strong_test} – with a score of {test_score}%. I am glad to see you succeed, and I believe that this achievement was because {reason}. {add_on}")
    except:
        return(f"\n\tThis semester, we have had three exams. You had the most success on the {strong_test} – with a score of {test_score}%. I am glad to see you succeed, and I believe that this achievement was because {reason}.")

def conclusion(student):
    '''Takes in the name of a student and outputs a personalized conclusion
    for their comment based on their grade in the class'''
    for i in INFO:
        if i[0] == student:
            if i[1] in ['A','A-']:
                indefinite_article = 'an'
                opinion = 'I look forward to your continued success during the remainder of the year.'
            else:
                indefinite_article = 'a'
                opinion = 'Your performance could use a little bump, whether it be in the form of one-on-one meetings with me during lunch or something as simple as more of a commitment to math.'

            return(f"I have appreciated having you in my class this semester. Overall, your grade was {indefinite_article} {i[1]}. {opinion}")


def comment(student):
    '''Takes in the name of a student and outputs a personalized semester comment'''
    intro = class_intro(student)
    strong = strong_trait(student)
    weak = weakness(student)
    learning = learning_outcome(student)
    test = strong_test(student)
    conc = conclusion(student)
    return intro + strong + weak + learning + test + conc

names = []
for i in INFO:
    names.append(i[0])

comments = []
for name in names:
    comments.append(comment(name))

for i in range(len(names)):
    with open(f'{names[i]}.txt','w') as f:
        f.write(comments[i])
```
